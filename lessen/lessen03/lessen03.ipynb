```python
import pandas as pd
```

```python
df = pd.read_csv('laptops.csv', sep=';')
df.head()
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
# df['Cpu_Company'] = df['Cpu'].
Cpu_Company = []
for i,row in df.iterrows():
    Cpu_Company.append(row['Cpu'].split(' ')[0])
    
df['Cpu_Company'] = Cpu_Company
df.head()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
Memory_Amount = []
for i,row in df.iterrows():
    Memory_Amount.append(row['Memory'].split(' ')[0])
    
df['Memory_Amount'] = Memory_Amount
df.head()
```

 ### Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
Memory_Amount = []
for i,row in df.iterrows():
    Memory_Amount.append(row['Memory'].split(' ')[0].replace('GB',''))
    
df['Memory_Amount'] = Memory_Amount
# df['Memory_Amount'].value_counts()
df.head()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
# Memory_Type = []
# for i,row in df.iterrows():
#     Memory_Type.append(row['Memory'].split(' ')[1:])
    
# df['Memory_Type'] = Memory_Type

# df['Memory_Type'] = df['Memory'].apply(labda x: ' '.join(x.split(' ')[1:]))
df['Memory_Type']= df['Memory'].apply(lambda x:' '.join(x.split(' ')[1:]))
df.head()
```

4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора

4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ


```python
df.pivot_table(index='Cpu_Company', columns='Ram',
               values='Weight', aggfunc='count',
               fill_value=0)
```

4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
df.pivot_table(index='OpSys', columns='Memory_Amount',
               values='Price_euros', aggfunc='mean',
               fill_value=0)
```

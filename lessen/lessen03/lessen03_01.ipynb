# Задание2

Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
Memory_Amount = []
for i,row in pf.iterrows():
    if 'GB' in row['Memory'].split(' ')[0]:
        Memory_Amount.append(int(row['Memory'].split(' ')[0].replace('GB', '')))
    elif 'TB' in row['Memory'].split(' ')[0]:
         Memory_Amount.append(int(float(row['Memory'].split(' ')[0].replace('TB', ''))*1024))
                                                    
pf['Memory_Amount'] = Memory_Amount
pf['Memory_Amount'].value_counts()
# pf['Memory_Amount'].dtype
```

```python
def mem(x):
    a=x.split(' ')[0]
    if a.endswith('GB'):
        res=a.replace('GB', '')
    elif a.endswith('TB'):    
        res=float(a.replace('TB', ''))*1024
    return int(res)    

pf['Memory_Amount2']=pf['Memory'].apply(mem)
# pf
```

# Задание3

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
pf['Memory_Type']=pf['Memory'].apply(lambda x:' '.join(x.split(' ')[1:]))
pf['Memory_Type'].value_counts()
```

```python
def mem1(x):
    listX=['SSD', 'HDD', 'Flash Storage','Hybrid']
    res=''
    for i in listX:
        if i in x:
            res+=' '+i
    return res
    

pf['Memory_Amount2']=pf['Memory'].apply(mem1)
pf['Memory_Amount2'].value_counts()
```

Процессоры от Samsung не изучайте

3.1 Постройте график в абсолютных величинах

```python
pf=df[~(df['Cpu_Company'] == 'Samsung')]
data = pd.crosstab(index = pf['Company'], columns = pf['Cpu_Company'], normalize='index').reset_index()#.sort_values(by=0, ascending=False)
# data.rename(columns = {0: 'no', 1: 'yes'}, inplace=True)
data
```

```python
# Строим столбчатую многорядную диаграмму
plt.figure(figsize =(8, 6))
# орпеделить оличество делении
n_ticks = np.arange(len(data['AMD']))
# определить сдвиг 
offset = 0.2
# определить ширину столбцов 
w =0.4
# добавляем сдвиг к количеству делении 
plt.bar(n_ticks - offset, data['AMD'], width=w)
plt.bar(n_ticks + offset, data['Intel'], width=w)

plt.title('род занятийб клиентов банкаб открывших депозит')
plt.xlabel('род зяняти')
plt.ylabel('Количество')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data['Company'], rotation = 30);
```

```python
np.arange(len(data['AMD']))
```

```python
plt.figure(figsize=(8,6))

plt.bar(data['Company'], data['AMD'])
plt.bar(data['Company'], data['Intel'], bottom=data['AMD'])

plt.legend(['AMD', 'Intel'])
plt.xticks(rotation= 70);
```

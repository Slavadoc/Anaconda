1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
delta_renovated = []
for i, row in df.iterrows():
    if 0 ==  row['yr_renovated']:
        delta_renovated.append(0)
    else:
        delta_renovated.append(row['yr_renovated'] - row['yr_built'])
        
df['delta_renovated'] = delta_renovated
df.head()
```

```python
df['lambda_renovated'] = df['yr_renovated'].apply(lambda x: 0  if x == 0 else df['yr_renovated'] - df['yr_built' ])
df.head()
```

#### 1.3 Создайте признаки года продажи, месяца продажи

```python
data = []
for i, row in df.iterrows():
    data.append(row['date'][:8])
data
# df.drop(columns = ['month_of_sale'], inplace=True)
df.head()
```

```python
years = []
for i in data:
    years.append(i[:4])

df['yr_of_sale']  =  years
df.head()
```

```python
month = []
for i in data:
    month.append(i[4] + i[5])

df['month_of_sale']  =  month
df.head()
```
